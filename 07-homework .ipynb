```python
#(1) 将所有中文简体汉字以及对应的unicode码的十进制数写到文本文件中，每行包含汉字和对应的unicode码的十进制数。
import unicodedata

# 创建一个空列表来存储汉字和对应的Unicode码
chinese_characters = []

# 定义汉字的Unicode范围，这里以基本汉字为例
start_code = 0x4E00  # 基本汉字的起始码
end_code = 0x9FFF      # 基本汉字的结束码

# 遍历Unicode码范围内的每个字符
for code in range(start_code, end_code + 1):
    char = chr(code)
    # 确保是汉字，而不是其他符号或空白字符
    if '\u4e00' <= char <= '\u9fff':
        # 获取字符和其Unicode码的十进制表示
        char_name = unicodedata.name(char)
        if char_name:  # 忽略那些没有名字的字符（如空白字符）
            chinese_characters.append((code, char, char_name))

# 将结果保存到文本文件
with open('chinese_characters.txt', 'w', encoding='utf-8') as f:
    for code, char, _ in chinese_characters:
        f.write(f'{code} {char}\n')

# 打印完成消息
print('所有中文简体汉字及其Unicode码已保存到chinese_characters.txt文件中。')
```

```python
#(2) 读取199801_source.txt，统计文本中出现的汉字频次，对汉字做降序排序；同时统计标点符号出现的频次，将单词词频和标点符号频次以json的形式分别写入两个文件。
import json
from collections import Counter

# 文件路径
file_path = r"C:\Users\86184\Desktop\文本分析\199801.txt"

# 打开文件并读取内容
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# 分离汉字和标点符号
chinese_chars = []
punctuation_marks = []
for char in text:
    if '\u4e00' <= char <= '\u9fff':
        chinese_chars.append(char)
    elif char in '，。；：！？《》（）【】「」『』“”‘’、—…' or char.isspace():
        punctuation_marks.append(char)

# 计算汉字和标点符号的频次
chinese_freq = Counter(chinese_chars)
punc_freq = Counter(punctuation_marks)

# 将结果转换为降序排序的字典
chinese_freq_sorted = dict(chinese_freq.most_common())
punc_freq_sorted = dict(punc_freq.most_common())

# 将结果写入JSON文件
with open('chinese_freq.json', 'w', encoding='utf-8') as f:
    json.dump(chinese_freq_sorted, f, ensure_ascii=False, indent=4)

with open('punc_freq.json', 'w', encoding='utf-8') as f:
    json.dump(punc_freq_sorted, f, ensure_ascii=False, indent=4)
```
